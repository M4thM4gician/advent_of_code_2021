# Advent of Code Challenges - Day 3

## Load Libraries

```python
import numpy as np
import pandas as pd

from pandas import Series,DataFrame
```

## Define Functions

```python
def binary_to_decimal(x):
    x = str(x)
    d = 0
    for n in np.arange(0,len(x)):
        d += (2**n)*int(x[-1-n])
    return(d)
```

## Load Data

```python
input_path = './input.txt'

with open(input_path) as f:
    lines = f.readlines()
f.close()

# Covert to pandas DataFrame
data = [[int(x) for x in line[0:12]] for line in lines]
df = DataFrame(data)
df.head()
```

## Puzzle 1 - We need more power, Captain!

```python
# Determine power consumption
# determine most common bit in each col to determine your gamma bit
# determine least common bit in each col to determine your epsilon bit
gamma_bit = ""
epsilon_bit = ""

for col in df.columns:
    col_sum = df[col].sum()
    if col_sum < (df.shape[0]/2):
        gamma_bit += "0"
        epsilon_bit += "1"
    else:
        gamma_bit += "1"
        epsilon_bit += "0"
print("gamma bit: " + gamma_bit)
print("epsilon bit: " + epsilon_bit + "\n")

gamma_dec = binary_to_decimal(gamma_bit)
epsilon_dec = binary_to_decimal(epsilon_bit)

print(f"gamma decimal: {gamma_dec}")
print(f"epsilon decimal: {epsilon_dec}" + "\n")

# Submit answer as product of gamma and epsilon decimal numbers
answer = gamma_dec*epsilon_dec
# 3687446
print(f"Answer: {answer}")
```

## Puzzle 2 - Just Breathe Normally

```python
# Determine the life support status
# Start with the bit in the first position and find the most common value (0 or 1, use 1 if there is a tie)
# Keep all bits that start with that number and move on to the second bit position
# Repeat the process of finding the most common bit in each position and filter remaining bits with the same bit values
# The last bit remaining is the 

oxy_gen_bit = ""
co2_scrub_bit = ""

oxy_df = df.copy()
co2_df = df.copy()

for col in oxy_df.columns:
    if oxy_df.shape[0] > 1:
        if oxy_df[col].sum() < oxy_df.shape[0]/2:
            bit_val = 0
        else:
            bit_val = 1
        oxy_df = oxy_df[oxy_df[col] == bit_val]
        oxy_gen_bit += str(bit_val)
    else:
        bit_val = oxy_df.iloc[0,col]
        oxy_gen_bit += str(bit_val)

for col in co2_df.columns:
    if co2_df.shape[0] > 1:
        if co2_df[col].sum() < co2_df.shape[0]/2:
            bit_val = 1
        else:
            bit_val = 0
        co2_df = co2_df[co2_df[col] == bit_val]
        co2_scrub_bit += str(bit_val)
    else:
        bit_val = co2_df.iloc[0,col]
        co2_scrub_bit += str(bit_val)

print("oxygen generator bit: " + oxy_gen_bit)
print("CO2 scrubber bit: " + co2_scrub_bit + "\n")

oxy_gen_dec = binary_to_decimal(oxy_gen_bit)
co2_scrub_dec = binary_to_decimal(co2_scrub_bit)

print(f"oxygen generator decimal: {oxy_gen_dec}")
print(f"CO2 scrubber decimal: {co2_scrub_dec}" + "\n")

# Submit answer as product of oxygen generator and CO2 scrubber decimal numbers
answer = oxy_gen_dec*co2_scrub_dec
# 4406844
print(f"Answer: {answer}")
```

### Verbose Loops

```python
oxy_gen_bit = ""

oxy_df = df.copy()

for col in oxy_df.columns:
    if oxy_df.shape[0] > 1:
        if oxy_df[col].sum() < oxy_df.shape[0]/2:
            bit_val = 0
        else:
            bit_val = 1
        print(f"Column Index: {col}")
        print(f"Count of '1s': {oxy_df[col].sum()}")
        print(f"Count of '0s': {oxy_df.shape[0] - oxy_df[col].sum()}")
        print(f"Bit value: {bit_val}")
        oxy_df = oxy_df[oxy_df[col] == bit_val]
        oxy_gen_bit += str(bit_val)
        print(f"oxy_gen_bit: {oxy_gen_bit}")
        print("")
    else:
        print(f"Column Index: {col}")
        print("One bit remains")
        bit_val = oxy_df.iloc[0,col]
        oxy_gen_bit += str(bit_val)
        print(f"Remaining Bit: {oxy_df.iloc[0]}")
        print(f"Bit value: {bit_val}")
        print(f"oxy_gen_bit: {oxy_gen_bit}")
        print("")
```
